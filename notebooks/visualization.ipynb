## Call all basic imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Project Results Summary

```python
results = pd.DataFrame({
    "Model": [
        "Fake News",
        "LIAR",
        "Twitter Bot"
    ],

    "Accuracy": [
        98.79,
        61.47,
        50.32
    ],

    "Precision": [
        99,
        63,
        50
    ],

    "Recall": [
        99,
        75,
        51
    ],

    "F1": [
        99,
        69,
        50
    ]
})

results
```

## LIAR Dataset Distribution

```python
liar_distribution = pd.DataFrame({
    "Class": ["True-ish", "False-ish"],
    "Count": [5752, 4488]
})

plt.figure(figsize=(8,5))
sns.barplot(data=liar_distribution,
            x="Class",
            y="Count")

plt.title("LIAR Dataset Class Distribution")
plt.show()
```

## Twitter Bot Distribution

```python
bot_distribution = pd.DataFrame({
    "Class": ["Bot", "Human"],
    "Count": [25018, 24982]
})

plt.figure(figsize=(8,5))
sns.barplot(
    data=bot_distribution,
    x="Class",
    y="Count"
)

plt.title("Twitter Bot Dataset Distribution")
plt.show()
```

## LIAR Confusion Matrix

```python
liar_cm = np.array([
    [400, 498],
    [291, 859]
])

plt.figure(figsize=(6,5))

sns.heatmap(
    liar_cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.title("LIAR Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

## Twitter Bot Confusion Matrix

```python
bot_cm = np.array([
    [2492, 2504],
    [2475, 2529]
])

plt.figure(figsize=(6,5))

sns.heatmap(
    bot_cm,
    annot=True,
    fmt="d",
    cmap="Reds"
)

plt.title("Twitter Bot Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

## Fake News Confusion Matrix

```python
fake_news_cm = np.array([
    [3811, 38],
    [59, 4092]
])

plt.figure(figsize=(6,5))

sns.heatmap(
    fake_news_cm,
    annot=True,
    fmt="d",
    cmap="Greens"
)

plt.title("Fake News Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

## Accuracy Comparison

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=results,
    x="Model",
    y="Accuracy"
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")

plt.show()
```

## Precision / Recall / F1 Comparison

```python
metrics_df = results.melt(
    id_vars="Model",
    value_vars=["Precision", "Recall", "F1"],
    var_name="Metric",
    value_name="Score"
)

plt.figure(figsize=(10,6))

sns.barplot(
    data=metrics_df,
    x="Model",
    y="Score",
    hue="Metric"
)

plt.title("Precision Recall and F1 Comparison")

plt.show()
```

## Dataset Size Comparison

```python
dataset_sizes = pd.DataFrame({
    "Dataset": [
        "Fake News",
        "LIAR",
        "Twitter Bot"
    ],
    "Samples": [
        40000,
        10240,
        50000
    ]
})

plt.figure(figsize=(8,5))

sns.barplot(
    data=dataset_sizes,
    x="Dataset",
    y="Samples"
)

plt.title("Dataset Size Comparison")

plt.show()
```
